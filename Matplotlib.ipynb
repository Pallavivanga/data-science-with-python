```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color='red')
```

```python
plt.plot(x,y,color='red',linewidth=5,linestyle='dotted')
```

```python
plt.xlabel('Weekdays')
plt.ylabel('temperature')
plt.title('line graph')
plt.plot(x,y,'+--',color='red')
```

```python
plt.xlabel('Weekdays')
plt.ylabel('temperature')
plt.title('line graph')
plt.plot(x,y,color='red',marker='+',markersize=3,alpha=0.2)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('temperature')
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
plt.legend(['max','min','avg'],loc='upper right',fontsize='large')
plt.grid()
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
plt.bar(days,max_t)
plt.xlabel('days')
plt.ylabel('temperature')
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
plt.xlabel('company')
plt.ylabel('revenue')
plt.legend('revenue')
```

```python
import numpy as np
y_val=np.arange(len(company))
```

```python
y_val
```

```python
plt.bar(y_val-0.2,revenue,width=0.4,label='revenue')
plt.xticks(y_val,company)
plt.legend()
plt.bar(y_val+0.2,profit,width=0.4,label='profit')
plt.legend()
```

```python
plt.legend(['revenue'])
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue,label='revenue',width=0.2)
plt.bar(company,profit,label='profit')
plt.legend()
```

```python
plt.barh(y_val-0.2,revenue,label='revenue')
plt.yticks(y_val,company)
plt.legend()
plt.barh(y_val+0.2,profit,label='profit')
plt.legend()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar,bins=(80,100,120,150),rwidth=0.75,color='g')
```

```python
plt.hist(blood_sugar,bins=(80,100,120,150),rwidth=0.75,color='g',histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women= [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 11]
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.85,color=['yellow','orange'],label=['men','women']
        )
plt.legend()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.85,color=['yellow','orange'],label=['men','women'],orientation='horizontal'
        )
plt.legend()
```

```python
#pie chart
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.1f%%',explode=[0,0,0.3,0,0],startangle=140,shadow=True)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,autopct='%0.1f%%',explode=[0,0,0.3,0,0],startangle=140,shadow=True)
plt.show()
plt.savefig('D:/data science with python/piechart.png',bbox_inches='tight',pad_inches=2,transparent=True)
```
