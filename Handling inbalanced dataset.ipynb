```python
#class count
count_class_0,count_class_1=df2.Churn.value_counts()
```

```python
count_class_0
```

```python
count_class_1
```

```python
df_class_0=df2[df2['Churn']==0]
df_class_1=df2[df2['Churn']==1]
```

```python
df_class_0.shape
```

```python
df_class_1.shape
```

```python
df_class_0.sample(count_class_1).shape
```

```python
df_class_0.sample(count_class_0).shape
```

```python
df_class_1.sample(count_class_1).shape
```

```python
df_class_0_under=df_class_0.sample(count_class_1)
df_test_under=pd.concat([df_class_0_under,df_class_1],axis=0)
print('Random under-sampling:')
print(df_test_under.Churn.value_counts())
```

```python
df_class_0_under.shape
```

```python
df_class_1.shape
```

```python
df_test_under.shape
```

```python
x=df_test_under.drop('Churn',axis=1)
y=df_test_under['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=15,stratify=y)
```

```python
y_train.shape
```

```python
y_train.value_counts()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
def ANN(x_train,y_train,x_test,y_test,loss):
    model=keras.Sequential([keras.layers.Dense(26,input_dim=26,activation='relu'),
                           keras.layers.Dense(15,activation='relu'),
                           keras.layers.Dense(1,activation='sigmoid')])
    model.compile(optimizer='adam',loss=loss,metrics=['accuracy'])
    model.fit(x_train,y_train,epochs=100)
    print(model.evaluate(x_test,y_test))
    y_pred=model.predict(x_test)
    y_pred=np.round(y_pred)
    print("classification report: \n", classification_report(y_test,y_pred))
    return y_pred
    
```

```python
y_pred=ANN(x_train,y_train,x_test,y_test,'binary_crossentropy')
```
