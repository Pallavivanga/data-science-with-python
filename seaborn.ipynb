```python
import seaborn as sns
```

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
```

```python
#catplot we can change the plot to any plot like violin plot ,boxplot,scatterplot by using kind='violine or box'
b=sns.load_dataset('tips')
```

```python
b.head()
```

```python
sns.catplot(x='day',y='total_bill',data=b)
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='violin')
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='box')
```

```python
sns.catplot(x='day',y='total_bill',data=b,kind='scatter')
```
